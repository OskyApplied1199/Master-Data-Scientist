1. Suppose that $f(x) = x^TQx$ where $Q \in R^{n\times n}, Q \geq 0$, $Q = Q^T$ . Show that $f(x)$ is convex $\forall x \in \mathbb{R}^n$.


Proof: To demonstrate that $f$ in convex, it suffices to prove that:
		$$\begin{equation}
			f( \lambda x_1 + (1-\lambda)x_2) \leq \lambda f(x_1) + (1-\lambda) f(x_2)
		\end{equation}$$
		We start by developing:
		\begin{align*}
			f( \lambda x_1 + (1-\lambda)x_2) &= (\lambda x_1 + (1-\lambda)x_2)^T Q (\lambda x_1 + (1-\lambda)x_2)\\
			&= (\lambda x_1^T + (1-\lambda)x_2^T) [Q \lambda x_1 + Q (1-\lambda)x_2]\\
			&=  \lambda x_1^T [Q \lambda x_1 + Q (1-\lambda)x_2] + (1-\lambda) x_2^T [Q \lambda x_1 + Q (1-\lambda)x_2]\\
			&= \lambda x_1^T Q \lambda x_1 + \lambda x_1^T Q (1-\lambda)x_2 +  (1-\lambda) x_2^T Q \lambda x_1 + (1-\lambda) x_2^T Q (1-\lambda)x_2\\
			&= \lambda^2 x_1^T Q x_1 + \lambda x_1^T Q (1-\lambda)x_2 +  (1-\lambda) x_2^T Q \lambda x_1 + (1-\lambda)^2 x_2^T Q x_2\\
			&= \lambda^2 f(x_1) + \lambda (1-\lambda)  [x_1^T Q x_2 +  x_2^T Q  x_1] + (1-\lambda)^2 f(x_2)\\
			&= \lambda^2 f(x_1) + 2 \lambda (1-\lambda) x_1^T Q x_2 + (1-\lambda)^2 f(x_2).
		\end{align*}  
        
The last because $x_1^T Q x_2$ is a scalar, then $x_1^T Q x_2 = x_2^T Q  x_1$, next if we subtract $\lambda f(x_1) + (1 \lambda) f(x_2)$ from the last part of equation, we get:
		\begin{align*}
			&\lambda^2 f(x_1) + 2 \lambda (1-\lambda) x_1^T Q x_2 + (1-\lambda)^2 f(x_2) - [\lambda f(x_1) + (1-\lambda) f(x_2)] \\
			&= \lambda^2 f(x_1) + 2 \lambda (1-\lambda) x_1^T Q x_2 + (1-\lambda)^2 f(x_2) - \lambda f(x_1) - (1-\lambda) f(x_2) \\
			&= (\lambda^2 - \lambda) f(x_1) + [(1-\lambda)^2 - (1-\lambda)] f(x_2) + 2 \lambda (1-\lambda) x_1^T Q x_2 \\
			&= \lambda (1 - \lambda) f(x_1) + \lambda(1 - \lambda) f(x_2) + 2 \lambda (1-\lambda) x_1^T Q x_2 \\
			&= \lambda (1 - \lambda) [f(x_1) + f(x_2) + 2 x_1^T Q x_2] \\
			&= \lambda (1 - \lambda) [x_1^T Q x_1 + 2 x_1^T Q x_2 + x_2^T Q x_2 ]\\
			&= \lambda (1 - \lambda) [(x_1+x_2)^T Q (x_1 + x_2)].
		\end{align*}  
        
Notice that $\lambda (1 - \lambda) > 0 \forall \lambda \in [0,1]$, also $Q \geq 0$, $Q = Q^T$ therefore $(x_1+x_2)^T Q (x_1 + x_2) >0$,   
$\therefore (1 - \lambda) [(x_1+x_2)^T Q (x_1 + x_2)] > 0$, in this way: $f( \lambda x_1 + (1-\lambda)x_2) \leq \lambda f(x_1) + (1-\lambda) f(x_2)$, is to say $f$ is convex. QED

2. Let it be $A  \in \mathbb{R}^{n\times n}$, $A = A^T$ . Consider $B = A + \alpha I$, where $I  \in \mathbb{R}^{n\times n}$ denotes the identity matrix and $\alpha \in \mathbb{R}^+$. Show that $B > 0$ for “big enough” values of $\alpha$.

Proof. To prove that $B > 0$, there is to show that $ x^T B x > 0$, for every non zero $x \in \mathbb{R}^n$, in other words, $x^T (A + \alpha  I) x > 0$. Expand:
\begin{align*}
	x^T(A + \alpha I) x &= x^TAx + x^T \alpha I x\\
	&= x^T A x + \alpha ||x||^2.
\end{align*}  

$A$ can be expressed in terms its eigenvalues using its orthogonal matrix to be diagonalized from fact to be a symmetric matrix, more over, the quadratic for $x^T A x$ can be simplified:
	\begin{equation*}
		x^T A x = \sum_{i=}^{n} \lambda_i (x_i)^2
	\end{equation*} 
	where $\lambda_i$ are the eigenvalues of $A$ and $x_i$ are the coordinates of x in the basis of eigenvectors of $A$.
 The eigenvalues provides of limits to $A$:
 $$\lambda_{min} ||x||^2 \leq x^T A x \leq \lambda_{max} ||x||^2 $$
 like $x^T A x \geq \lambda_{min} ||x||^2$, then $x^T A x + \alpha ||x||^2 \geq \lambda_{min} ||x||^2 + \alpha ||x||^2$.  
 $$ x^T A x + \alpha ||x||^2 \geq ||x||^2 [\lambda_{min} + \alpha] $$
 $$ x^T B x \geq ||x||^2 [\lambda_{min} + \alpha]$$
 $x^T B x > 0$ for any $x \in \mathbb{R}^n$, then must be $[\lambda_{min} + \alpha] > 0$. 
 Hence, solving for $\alpha$,
 $$ \alpha > -\lambda_{min} $$.
 Therefore for any $\alpha$ that is greather than $-\lambda_{min}$, $x^T B x > 0$, is to say, $B$ is definide positive. QED.

```python
#Gradiente descendiente metodo
def gradesc(learning_rate, iterar , gradf, x):
    x = np.asarray(x, dtype=np.float64)
    history_x = [x]
    for _ in range(iterar):
       grad = gradf(x)
       x = x - learning_rate*grad
       history_x.append(x)
    return history_x

#Funcion y grad a)
def fa(x):
    a = 0
    n = len(x)
    for i in range(n):
        a += (x[i]-c[i])**2
    return a

def gfa(x):
    y = np.zeros(len(x))
    for i in range(len(x)):
        y[i] = 2*(x[i] - c[i])
    return  y

c = np.array([1,1,1])

#Funcion y grad b)
def fb(x):
    n = len(x)
    suma=0
    for i in range(n-1):
        suma += (100*(x[i+1] - x[i])**2 - (x[i]-1)**2)
    return suma

def gfb(x):
    x = np.asarray(x, dtype=np.float64)
    n = len(x)
    y = np.zeros(n, dtype=np.float64)
    
    # Gradiente para el primer elemento
    y[0] = -400 * x[0] * (x[1] - x[0]**2) - 2 * (1 - x[0])
    
    # Gradiente para los elementos intermedios
    for i in range(1, n - 1):
        y[i] = 200 * (x[i] - x[i - 1]**2) - 400 * x[i] * (x[i + 1] - x[i]**2) - 2 * (1 - x[i])
    
    # Gradiente para el último elemento
    y[n - 1] = 200 * (x[n - 1] - x[n - 2]**2)
    return y

#Funcion y grad c)
def fc(x):
    beta = 1
    n  =len(x)
    c = 0
    k = 0
    for i in range(1,n+1):
        c = 0
        for j in range(1, n+1):
            c += (j**i + beta)*((x[j-1]/j)**i-1)
        k += c**2
    return k

def gfc(x):
    n = len(x)
    X = np.zeros(n)
    for k in  range(1, n+1):
        m=0
        for i in range(1, n+1):
            c = 0
            for j in range(1, n+1):
                c += (j**i+1)*((x[j-1]/j)**i-1)
            m += i*(k**i+1)*c
        X[k-1] = 2*m/k
    return X
```

```python
#Gradiente numerico
def gradf(x, f, h=1e-4) :
    n = len(x)
    grad = np.zeros(n)
    
    for i in range(n):
        x_i = np.copy(x)
        x_i[i] += h
        grad[i] = (f(x_i) - f(x)) / h
    return grad
# Metodo para grandiente numerico
def gradescnum(learning_rate, iterar , f, x):
    x = np.asarray(x, dtype=np.float64)
    history_x = [x]
    for _ in range(iterar):
       grad = gradf(x,f)
       x = x - learning_rate*grad
       history_x.append(x)
    return history_x
```
