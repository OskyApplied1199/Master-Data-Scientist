7. Show that of all rectangular parallelepipeds with a given surface, the cube has the greatest volume.  
Proof. Let be an rectangular parallelepiped with $V=abc$, and surface $S=2[ac + ab + bc] = k$.  
For this exercise, we apply the multipliers Lagrange method: $\nabla F = \lambda \nabla g$,  where $f = V$ and $g = S$.  
$$ 
\begin{align*}
bc &= \lambda 2[c+b] &\rightarrow bc = 2\lambda c + 2 \lambda b \\
ac &= \lambda 2[a+c] &\rightarrow ac = 2\lambda a + 2 \lambda c \\
ab &= \lambda 2[a+b] &\rightarrow bc = 2\lambda a + 2 \lambda b \\
2[ac + ab + bc] &= k
\end{align*}
$$
Solve this equations system, concluse that: $a = b$ and  $b = c$, by transitivy identy $a=c \therefore V=a^3$, the parallelepiped is a cube.

9. a)Make a script in Python to approximate their gradients and Hessians,evaluated in a particular point x, utilizing finite differences.

```python
#Script for gradient
import numpy as np

def gradf(x, f, h=1e-6) :
    n = len(x)
    grad = np.zeros(n)
    
    for i in range(n):
        x_i = np.copy(x)
        x_i[i] += h
        grad[i] = (f(x_i) - f(x)) / h
    return grad

#Script for Hessiana
def hessian(f, x, h=1e-6):
    n = len(x)
    hess = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            x_ij_h = np.copy(x)
            x_ij_h[i] += h
            x_ij_h[j] += h
            f_ij_h = f(x_ij_h)
            
            x_i_h = np.copy(x)
            x_i_h[i] += h
            f_i_h = f(x_i_h)
            
            x_j_h = np.copy(x)
            x_j_h[j] += h
            f_j_h = f(x_j_h)
            
            f_0 = f(x)
            
            hess[i, j] = (f_ij_h - f_i_h - f_j_h + f_0) / (h ** 2)
    return hess
```

10. Consider Newton’s method for function minimization given by the iterative formula:

$$ \mathbf{x}_{k+1} = \mathbf{x}_k - \left[\nabla^2 f(\mathbf{x}_k)\right]^{-1} \nabla f(\mathbf{x}_k), \quad k = 0, 1, 2, \dots, K$$
